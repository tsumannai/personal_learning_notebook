## Probability Denisities

If probability of a real valued variable $x$ falling in interval $(x,x+\delta{x})$ is given by $p(x)\delta{x}$ for $\delta{x}\rightarrow0$ then $p(x)$ is called probability density over $x$.

Probability of $x$ between $(a,b)$ will be given by:

$$p(x \in (a,b)) = \int_{a}^{b} p(x) \,dx $$

- Probabilities cannot be negative and be on a real axis.
- Total probabilites cannot exceed 1.

$$p(x) \ge 0$$
$$\int^{\infty}_{-\infty}p(x)dx = 1$$

For nonlinear transformations:

If $x = g(y)$ then $f(x)$ becomes $f(y) = f(g(y))$.

For probability denisity $p_x(x)$ that corresponds to another density $p_y(y)$.

$$p_y(y) = p_x(x)|\frac{dx}{dy}|$$
$$ = p_x(g(y))|g'(y)|$$

The maximum of a probability density function would then depend on the variable/transformation.

For several continuous variables denoted collectively by vector $x$, a joint probability density $p(x) = p(x_1, ..., x_D)$ such that the probability of $x$ falling by $\delta{x}$ contaiing point x is given by $p(x)\delta{x}$. Thus must satisfy:

$$p(x) \ge 0$$
$$\int{}p(x)dx = 1$$

Sum and product rules and Bayes' theorem apply equally to probability denisites. If $x$ and $y$ are real variables:

$$p(x) = \int{}p(x,y)dy$$
$$p(x,y) = p(y|x)p(x)$$
