# Probability Theory

The purpose of probability theory is so provide a consistent framework for quantification and manipulation of uncertainty. When combined with decision theory, it allow for optimal predictions given information available to us.

For two random variables $X$ and $Y$.

$X$ can take any values $x_i -> x=M$ and $Y$ can take values $y_i -> y=L$.

Number of trials where  $X = x_i$ and $Y = y_j$ is $n_{ij}$

For $N$ trials where both variables are sampled by $n_{ij}$

For number of trials of $X = x_i$ irrespective of $Y$ value $c_i$. <br>
For number of trials of $Y = y_j$ takeirrespective of $X$ value $r_j$.



$$p(X = x_i, Y = y_i) = \frac{n_{ij}}{N}$$

Joint probability is expressed as a fraction of total number of points.

By columns, probability of $X = x_i$ irrespective of $Y$:
$$p(X = x_i) = \frac{c_i}{N}$$

$$c_i = \sum_jn_{ij}$$

Thereby we have the sum rule of probability, or marginal probability.

$$p(X = x_i) = \sum_{j=1}^Lp(X = x_i, Y = y_j)$$

If we consider instances where $X = x_i$ written then we can infer the conditional probability of $Y = y_i$.

$$p(Y = y_i|X = x_i) = \frac{n_{ij}}{c_i}$$

$$p(X = x_i,Y = x_j) = \frac{n_{ij}}{N} = \frac{n_{ij}}{c_i} * \frac{c_i}{N}$$
$$= p(Y = y_j|X = x_i)p(X = x_i)$$ 

### Rules of Probability

$$sum rule -> p(X) = \sum_Yp(X,Y)$$
$$product rule -> p(X,Y) = p(Y|X)p(X)$$

Since $$p(X,Y) = p(Y,X)$$ we can obtain Bayes theorem from:
$$p(Y|X)p(X) = p(X|Y)p(Y)$$
$$p(Y|X) = \frac{p(X|Y)p(Y)}{p(X)}$$

$$p(X) = \sum_Yp(X|Y)p(Y)$$

Probability of picking red or blue box.
$$p(B=r) = \frac{4}{10}$$
$$p(B=b) = \frac{6}{10}$$

What's the probability of choosing an apple?

$$p(F=a) = p(F=a|B=r)p(B=r) + p(F=a|B=b)p(B=b)$$
$$ = \frac{1}{4} * \frac{4}{10} + \frac{3}{4} * \frac{6}{10} = \frac{11}{20}$$

What's the probability of choosing from the red box if we picked an orange?

$$p(B=r|F=o) = \frac{p(F=o|B=r)p(B=r)}{p(F=o)}$$

$$ \frac{3}{4} * \frac{4}{10} * \frac{20}{9} = \frac{2}{3}$$
